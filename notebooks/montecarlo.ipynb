```python
import tensorflow as tf
import matplotlib.pyplot as plt

num_points = 1000

x = tf.random.uniform([num_points], minval=-1, maxval=1)
y = tf.random.uniform([num_points], minval=-1, maxval=1)

inside_circle = tf.cast(x**2 + y**2 <= 1, tf.bool)

x_inside = tf.boolean_mask(x, inside_circle)
y_inside = tf.boolean_mask(y, inside_circle)
x_outside = tf.boolean_mask(x, ~inside_circle)
y_outside = tf.boolean_mask(y, ~inside_circle)

# Calculando a estimativa de Ï€
pi_estimate = 4 * tf.reduce_sum(tf.cast(inside_circle, tf.float32)) / num_points
print(f"Estimativa de Ï€: {pi_estimate.numpy()}")

plt.figure(figsize=(6, 6))
plt.scatter(x_inside, y_inside, color='blue', label='Dentro do CÃ­rculo', s=1)
plt.scatter(x_outside, y_outside, color='red', label='Fora do CÃ­rculo', s=1)

circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=2)
plt.gca().add_artist(circle)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title(f"Estimativa de Ï€ com MÃ©todo de Monte Carlo: {pi_estimate.numpy():.4f}")
plt.legend()
plt.axis("equal")
plt.show()
```
